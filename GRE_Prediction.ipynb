```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('/content/Admission_Predict_Ver1.1.csv')
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.duplicated().sum()
```

```python
dataset.drop(columns='Serial No.',inplace=True)
```

```python
dataset.head()
```

```python
X=dataset.iloc[:,0:-1]
y=dataset.iloc[:,:-1]
print(X)
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()

X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(7,activation='relu',input_dim=7)) #input layer
model.add(Dense(7,activation='relu')) #input layer
model.add(Dense(1,activation='linear')) #output layer
```

```python
model.summary()
```

```python
model.compile(loss='mean_squared_error',optimizer='Adam')
```

```python
history = model.fit(X_train_scaled,y_train,epochs=100,validation_split=0.2)
```

```python
y_pred=model.predict(X_test_scaled)
```
